```python
import xarray as xr
import numpy as np
import pandas as pd
import glob
file_pattern = "F:\CDA-LAB\dataset\HadISST_sst.nc.gz"
file_paths = glob.glob(file_pattern)
ds = xr.open_mfdataset(file_paths)
print(ds)
```

```python
import matplotlib.pyplot as plt
p = ds['sst']
regional_sst = p.sel(latitude = slice(5,-5),longitude = slice(-170,-120))
sst_nino34_mean = regional_sst.mean(dim = ['latitude','longitude'])
sst_nino34_mean
```

```python
 import numpy as np
from scipy import stats
```

```python
t1 = sst_nino34_mean.sel(time = slice("1901-01-01","1930-12-31"))
t2 =  sst_nino34_mean.sel(time = slice("1995-01-01","2024-12-31"))
t1_mean = t1.mean(dim = 'time')
t2_mean = t2.mean(dim = 'time')
t_statistic, p_value = stats.ttest_ind(t1, t2, equal_var = False)
print(t_statistic)
```

```python
f_statistic, p_value = stats.f_oneway(t1,t2)
print(f_statistic)
```
